## Step 1: Import Libraries

Libraries are tools that help us do tasks easily.

- `pandas` → work with tables (data)
- `numpy` → work with numbers
- `matplotlib` → create graphs


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```


## Step 2: Create a Simple Dataset

A dataset is like a table (rows and columns).

We will create data for 5 patients.


```

data = {
    'patient_id': [1, 2, 3, 4, 5],
    'heart_rate': [72, 85, 78, 90, 76],
    'temperature': [36.6, 37.0, 38.2, 37.5, 36.9]
}

df = pd.DataFrame(data)

df
```


## Step 3: Look at the Data

We check:
- First rows
- Data types


```
df.head()
```

```
df.info()
```


## Step 4: Understand the Data

We use `.describe()` to see:
- average (mean)
- minimum value
- maximum value


```
df.describe()
```


## Step 5: Calculate Specific Values

Example: average heart rate


```
df['heart_rate'].mean()
```


## Step 6: Find Maximum and Minimum

This helps us understand extreme values.


```

df['heart_rate'].max()
```

```

df['heart_rate'].min()
```


## Step 7: Create a Graph

Graphs help us visualize the data.


```

plt.plot(df['patient_id'], df['heart_rate'])
plt.xlabel('Patient ID')
plt.ylabel('Heart Rate')
plt.title('Heart Rate of Patients')
plt.show()
```


## Step 8: Add New Column

We can create new data from existing data.

Example: Check if patient has high temperature (>37.5)


```

df['high_temp'] = df['temperature'] > 37.5
df
```
