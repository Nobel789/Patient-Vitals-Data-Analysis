## Step 1: Import Libraries


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```


## Step 2: Create a Larger Dataset

Now we simulate data for multiple patients over time.


```

np.random.seed(0)

n = 50

df = pd.DataFrame({
    'patient_id': np.random.randint(1, 6, n),
    'heart_rate': np.random.randint(60, 100, n),
    'temperature': np.random.normal(37, 0.5, n),
    'ward': np.random.choice(['A', 'B'], n)
})

df.head()
```


## Step 3: Basic Exploration


```
df.shape
```

```
df.describe()
```


## Step 4: Filter Data

Example: Patients with high heart rate (>90)


```

high_hr = df[df['heart_rate'] > 90]
high_hr
```


## Step 5: Group Data

We calculate average heart rate per ward.


```

df.groupby('ward')['heart_rate'].mean()
```


## Step 6: Missing Values

We create some missing values and then fix them.


```

df.loc[0:5, 'temperature'] = np.nan
df.isna().sum()
```

```

df['temperature'] = df['temperature'].fillna(df['temperature'].mean())
df.isna().sum()
```


## Step 7: Better Visualization

Histogram of heart rate


```

plt.hist(df['heart_rate'])
plt.xlabel('Heart Rate')
plt.ylabel('Count')
plt.title('Distribution of Heart Rate')
plt.show()
```


## Step 8: Compare Groups

Boxplot of heart rate by ward


```

df.boxplot(column='heart_rate', by='ward')
plt.title('Heart Rate by Ward')
plt.suptitle('')
plt.show()
```
